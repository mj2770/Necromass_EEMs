```python
import pandas as pd
import re
from plotnine import *
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import kruskal
```

```python
# Directory containing the CSVs
base_dir = Path("/content/drive/My Drive/LBNL Research/Romy/ENIGMA/Enrichments/Fresh vs Recycled necromass/EEMs")

# Find all CSV files (non-recursive). Use rglob("*.csv") if you want recursive.
csv_files = sorted(base_dir.glob("*.csv"))

# Get names before ".csv"
names = [p.stem for p in csv_files]

# Put into a DataFrame
df = pd.DataFrame({"name": names})

# Split by "_" into multiple columns
# expand=True creates separate columns; n controls max splits (optional)
parts = df["name"].str.split("_", expand=True)

# Give split columns names like part1, part2, ...
parts.columns = [f"part{i+1}" for i in range(parts.shape[1])]

# Combine
df = pd.concat([df, parts], axis=1)
df = df[~df["name"].str.startswith("._", na=False)].reset_index(drop=True)
df.to_csv("/content/drive/My Drive/LBNL Research/Romy/ENIGMA/Enrichments/Fresh vs Recycled necromass/EEMs/EEMs_metadata.csv",index=False)
```
